```python
import requests as req
from bs4 import BeautifulSoup
```

```python
url = "https://pypi.org/project/beautifulsoup4/"
response = req.get(url)
print(response)
```

```python
raw_html = response.text
raw_html
```

```python
beauty_html = BeautifulSoup(raw_html, "html.parser")
beauty_html
```

```python
beauty_html.title
```

```python
beauty_html.title.title
```

```python
beauty_html.h1
```

```python
beauty_html.h1.string
```

```python
beauty_html.nav
```

```python
h2 = beauty_html.find_all('h2')
h2
```

```python
anchor = beauty_html.find_all('a')
anchor
```

```python
for a in anchor:
    print(a)
```
